```python
import networkx as nx
import matplotlib.pyplot as plt

# Kagome unit cell
num_sites = 12

# Generate graph of kagome unit cell
# Start by defining all the edges
G = nx.Graph()
G.add_nodes_from(range(num_sites))
node_map = [2,4,10,13,11,5,1,7,12,14,8,3]
edge_list = [
    (0, 1, 0),
    (1, 2, 0),
    (2, 3, 0),
    (3, 4, 0),
    (4, 5, 0),
    (5, 0, 0),
    (0, 6, 1),
    (1, 6, 1),
    (1, 7, 1),
    (2, 7, 1),
    (2, 8, 1),
    (3, 8, 1),
    (3, 9, 1),
    (4, 9, 1),
    (4, 10, 1),
    (5, 10, 1),
    (5, 11, 1),
    (0, 11, 1),
]
colors = ('gray', 'green')
for e in edge_list:
    G.add_edge(e[0],e[1],color=colors[e[2]],width=e[2]+1)
pos = nx.spring_layout(G)
plt.figure()
nx.draw(
    G, pos, edge_color=nx.get_edge_attributes(G,'color').values(), width=list(nx.get_edge_attributes(G,'width').values()), linewidths=1,
    node_size=500, node_color='pink', alpha=0.9,
    labels={node: node_map[node] for node in G.nodes()}
)
nx.draw_networkx_edge_labels(
    G, pos,
    edge_labels={(0, 1): 'm', 
                 (1, 2): 'n',
                 (2, 3): 'o',
                 (3, 4): 'p',
                 (4, 5): 'q',
                 (5, 0): 'r',
                 (0, 6): 'a',
                 (1, 6): 'b',
                 (1, 7): 'c',
                 (2, 7): 'd',
                 (2, 8): 'e',
                 (3, 8): 'f',
                 (3, 9): 'g',
                 (4, 9): 'h',
                 (4, 10): 'i',
                 (5, 10): 'j',
                 (5, 11): 'k',
                 (0, 11): 'l'},
    font_color='red'
)
plt.axis('off')
plt.savefig('kagome.png')
plt.show()
```
