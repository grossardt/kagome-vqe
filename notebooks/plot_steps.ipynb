# Plot of the evolution of the ground state energy with iterations of the VQE algorithm

The lower green line shows the target ground state energy, the upper green line shows the accuracy
goal of a less than 1% error.

```python
import matplotlib.pyplot as plt

file1 = '20230315-140403-kagome-new.txt' # with noise
file2 = '20230315-141023-kagome-new.txt' # without noise

data = []
for file in (file1, file2):
    file_data = []
    with open(file, 'r') as f:
        line = f.readline()
        while line:
            if line[:4] == 'STEP':
                file_data.append(float(line[33:]))
            line = f.readline()
    f.close()
    data.append(file_data)

plt.plot(range(len(data[0])), data[0], 'r.', label='with noise')
plt.plot(range(len(data[1])), data[1], 'b--', label='without noise')
plt.axhline(y=-18, color='g', linestyle='-')
plt.axhline(y=-18*0.99, color='g', linestyle='-')
plt.legend()
plt.show()
```
